```python
# Read the file with the detected encoding and write it back in UTF-8
import pandas as pd


file_path = 'data/data.csv'
# Read the file with the detected encoding
data = pd.read_csv(file_path, encoding='GB2312')

# Save the file in UTF-8 encoding
output_path = 'data/data_utf8.csv'
data.to_csv(output_path, index=False, encoding='utf-8')

output_path
```
